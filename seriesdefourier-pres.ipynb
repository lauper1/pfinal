# Análisis de Fourier para Funciones definidas en Intervalos

Queremos ser capaces de poder analizar una señal en términos de su serie de Fourier.

La motivación de esto es que la transformada de Fourier sirve para obtener las frecuencias qe componen a cierta señal. En el laboratorio, los datos que se obtienen tienen una incertidumbre asociada al aparato de medición que arroja dichos datos. Dicha incertidumbre introducira el carácter de intervalo a las series de datos que analicemos.

De esta manera el dato $y_n$ se expresará como un intervalo $$[y_n-\delta,~y_n+\delta],$$ donde $\delta$ es la incertidumbre del aparato de medición.

La Transformada Discreta de Fourier está definida como:
\begin{equation}
X(k)=\Sigma_{j=1}^{N} x(j) w_N^{(j-1)(k-1)} 
\end{equation}


donde $w_N=e^{-2\pi i N}$


El desarrollo del trabajo está dividido en dos partes:

En la primera desarrollamos la Transformada Discreta de Fourier (DFT) para números flotantes con el fin de tener una idea de las operaciones necesarias para llevar la DFT para funciones definidas en intervalos.

Posteriormente se desarrolla el proceso de la DFT para intervalos, para lo cual (basados en la primera parte) es necesario cubrir tres puntos:




1) Generar y leer una señal, cuadrado, diente de sierra, o una señal arbitraria periódica definida en intervalos.

3) Realizar las sumas y productos definidas para  $X(k)$. 

4) Visualizar los resultados.

## 1. DFT para flotantes.

Primero se genera una señal de prueba aleatoria, con dos componentes correspondientes a dos frecuencias. Las frecuencias utilizadas son de 50 y 120. 

La idea es que después de aplicar la DFT se recuperen dichas frecuencias.

Adicionalmente la función que genera la prueba añade un ruido aleatorio a la señal, simulando lo que se obtendría en el laboratorio.

```julia
function sin_ruido(F, L)
    y=zeros(L,2)
    T=1/F
    x=zeros(L)
    t=zeros(L)
    D={}
    for i=1:L
        t[i]=i*T
        
        x[i]= 0.7*sin(2*pi*50*t[i])+sin(2*pi*120*t[i])
    end
    p=(F,L)
    y= x+rand(length(x))
    D=push!(D,p,t,y)
end
```

Esta función tiene dos parámetros, $F$ y $L$, $F$ es la frecuencia a la cual se lleva a cabo la meddición y $L$ es el número de mediciones que se llevan a cabo. Al final el output es una arreglo de 3 elementos. 

- El primer elemento es un vector que contiene dos entradas, la primera entrada es la frecuencia de medición y la segunda es el número de mediciones. 

- La segunda entrada del output es un arreglo de *L* números que contiene la escala de tiempo de la señal, el eje x.

- La tercera entrada del output es un arreglo de *L* números que contiene los datos del eje y de la señal. 

```julia
F=1000
L=1000
D=sin_ruido(F, L)
```

La señal resultante es la siguiente:

```julia
using PyPlot
plot(D[2],D[3])
```

Ahora defino una función auxiliar para la DFT, que lo que hace es caclular el elemento k-ésimo $$\Sigma_{j=1}^{N} x(j) w_N^{(j-1)(k-1)} $$

```julia
function TFY(D, k, N)
    y=D[3]
    w=(cos(-2*π*(k-1)/N)-im*sin((-2*π*(k-1)/N)))
    sum([y[n]*w^n for n=1:N])
end
```

Y en base a esta función se define la DFT, que ordena en una serie de datos a cada uno de los elementos calculados por TFY.

```julia
function DFT(D)
    y=D[3]
    N=length(y)
    FY={}
    for k=1:N
        FY=push!(FY, TFY(D, k, N))
    end 
FY     
end
```

Ahora, se aplica la DFT al conjunto de datos de la señal de prueba.

```julia
DFT(D)
```

Para obtener la gráfica que nos dirá donde se encuentran las componentes en frecuencia de la señal de prueba, es necesario sacar la norma de cada uno de los elementos del arreglo anterior. 

```julia
function norma_dft(D)
    N=zeros(length(D))
    for j in 1:length(D)
        N[j]=abs(D[j])
    end
    N
end
```

Ya con esta función se puede graficar el resultado de la DFT.

```julia
function graficar_DFT(D)
    F=norma_dft(DFT(D))
    f = D[1][1]*linspace(0,1,length(F))
    PyPlot.plot(f,F/D[1][2])
    PyPlot.xlim(0,150)
    PyPlot.ylim(0,0.6)
end
```

```julia
graficar_DFT(D)
```

Se observa que los picos de la transformada de Fourier se encuentran en 50 y 120, que, como ya se mencionó, son las frecuencias que componen a nuestra señal de prueba.

## 2. DFT para intervalos.

Al realizar la DFT utilizamos las raíces  n-ésimas de la unidad en el plano complejo. De modo que a fin de obtener la DFT para intervalos debemos ser capaces de tener definidos Intervalos Complejos. Para esto realizamos un módulo en el que se define el tipo IC el cual esta formado por un objeto R (parte real) y otro I (parte imaginaria). La idea es que en cada uno de estos objetos pueda haber un intervalo.

Así mismo se definieron las operaciones entre intervalos e intervalos complejos necesarias para realizar la DFT.

```julia
using Intervalos  #este modulo ya incluye intervalos complejos
```

```julia
R=Intervalo(0.1,1)
I=Intervalo(0.1,1)
a=IC(R, I)
```

Generar una señal a analizar cuya serie de datos  "y" estén formados por intervalos.  Para esto generaremos los datos en la forma dada para flotantes  y luego la entrada se mandará a su forma como intervalo.

```julia
F=1000
L=1000
D=sin_ruido(F, L);
```

La siguiente función expresará la entrada "y" de los datos (D[3]) como intervalo.

```julia
function conversion_inter(D)
    R={}
    S=Intervalo[]
    for j=1:length(D[3])
        S=push!(S,Intervalo(D[3][j]))
    end
    R=push!(R,D[1],D[2],S)
end
```

Esto **intervaliza** los datos correspondientes a la variable "y". Dejando iguales los parámetros de medición y el tiempo.

```julia
D2=conversion_inter(D)
```

A continuación mostramos una de la entradas del conjunto de datos, ya como intervalo.

```julia
D2[3][40] 
```

Lo siguiente es visualizar estos resultados en una gráfica. Primero es necesario sacar la norma de cada uno de los intervalos complejos en el resultado de la DFT. Esto se hace con la siguiente función auxiliar:

```julia
function norma_inter(y)
    r={}
    for n=1:length(y)
        r=push!(r,norm(y[n]))
    end
    r
end
```

Sacar esta norma hace que el resultado esté en términos de intervalos reales. Para visualizar esto se graficará la envolvente de estos intervalos. Esto es se graficará la curva que une a todas las cotas superiores del intervalo resultado y luego la gráfica de las cotas inferiores.

```julia
function cota_a(A)
    B=zeros(length(A))
    for n=1:length(A)
        B[n]=A[n].a
    end
    B
end
function cota_b(A)
    B=zeros(length(A))
    for n=1:length(A)
        B[n]=A[n].b
    end
    B
end
```

```julia
function graficar_cotas(D)
    F=norma_inter(DFT(D))
    Fa=cota_a(F)
    Fb=cota_b(F)
    f = D[1][1]*linspace(0,1,length(F))
    PyPlot.plot(f,Fa/D[1][2])
    PyPlot.plot(f,Fb/D[1][2])
    PyPlot.xlim(0,150)
end
```

Esta última función genera intervalos entre los que se hallan las normas de Transformada Discreta de Fourier. Al graficarla debemos obtener una gráfica muy parecida a la que obtuvimos para flotantes, donde recuperabamos las frecuencias introducidas a la señal de  prueba, pero definida en un rango dado por la longitud de los intervalos.

```julia
graficar_cotas(conversion_inter(D))
```

La curva superior e inferior coinciden casi exactamente punto a punto. Esto se debe a que al momento de **intervalizar** la señal el proceso de intervalizacion lo realiza con un rango dado por los dos flotantes mas cercanos al valor dado por la señal, que en realidad es un valor muy cercano a cero.

Hasta este punto podemos decir que los resultados obtenidos  para calcular la DFT al trabajar con flotantes o con intervalos coinciden bastante.

¿Qué otro uso podemos darle a nuestro programa además de la comprobación de los cálculos con flotantes?

Pensamos que el trabajar con intervalos es semejante a la manera en que se manejan las incertidumbres en en análisis de datos experimentales.

Cuando se miden valores de un osciloscopio la exactitud de los mismos están limitados a la mínima escala que el osciloscopio es capaz de medir. En este sentido, a cada medida es posible asignarle una incertidumbre.

A continuación presentamos un breve ejemplo de como se trataría este caso escogiendo la incertidumbre y posteriormente calculando la DFT con los intervalos definidos por la misma.

Esta primera función **intervaliza** la señal escogiendo la norma del intervalo dado, que debe de ser la mínima escala de aparato de medición (tanto en x como en y).

```julia
function conversion_inter_del(D, d, e)   
    R={}
    S=Intervalo[]
    U=Intervalo[]
    for j=1:length(D[2])
        U=push!(U,Intervalo(D[2][j]-d,D[2][j]+d))
    end
    for j=1:length(D[3])
        S=push!(S,Intervalo(D[3][j]-e,D[3][j]+e))
    end
    R=push!(R,D[1],U,S)
end
```

Ya con los datos en términos de intervalos, se procede realizar la gráfica.

```julia
F=norma_dft(DFT(D))
D2=conversion_inter_del(D,0.001,0.01)
R=norma_inter(DFT(D2));
```

```julia
function graficar_cuadros(F,D2,R)
    f = D[1][1]*linspace(0,1,length(F))
    PyPlot.plot(f,F/D[1][2])
    PyPlot.xlim(0,150)
    #PyPlot.ylim(0,1)
    for k=1:length(D[3])
        PyPlot.fill_between([D[1][1]*float64(D2[2][k].a),D[1][1]*float64(D2[2][k].b)],float64(R[k].a)/D[1][2],float64(R[k].b)/D[1][2])
    end
end
```

```julia
graficar_cuadros(F,D2,R)
```


Los dos cuadrados azules que se encuentran en las frecuencias 50 y 120 son el resultado buscado. Contienen la información de las frecuencias e amplitudes relativas que componen a la señal. La información que brindan no es una aproximación al valor real sino un intervalo dentro del cual se encuentra el resultado. Esta es la parte que es de suma importancia en el análisis de datos, donde no se tiene un valor definitivo sino un intervalo de error dentro del cual está el resultado.

```julia
function sierra_ruido(F, L)
    y=zeros(L,2)
    T=1/F
    x=zeros(L)
    t=zeros(L)
    D={}
    for i=1:L
        t[i]=i*T
        
        x[i]=2/π *sum([((-1)^k*sin(2*pi*k*t[i])/k) for k in 1:2:150])
    end
    p=(F,L)
    y= x#+rand(length(x))
    D=push!(D,p,t,y)
end
```

```julia
S=sierra_ruido(1000,1000)
```

```julia
plot(S[2],S[3])
```

```julia
F=norma_dft(DFT(S))
D2=conversion_inter_del(S,0.001,0.01)
R=norma_inter(DFT(D2));
```

```julia
graficar_cuadros(F,D2,R)
```
